```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
```

```python
reviews = pd.read_csv('./data/reviews_preprocessed.csv')

print(reviews.shape)
```

```python
path = '/Library/Fonts/NanumBarunGothic.otf'
font_name = fm.FontProperties(fname=path, size=50).get_name()
plt.rc('font', family=font_name)
```

### 전체 리뷰 평점 히스토그램 

```python
reviews.hist(column = 'total_star', grid=False, figsize=(12,8), color='#86bf91')

plt.title("")
plt.xlabel("평점", labelpad=20, weight='bold', size=20)
plt.ylabel("리뷰 수", labelpad=20, weight='bold', size=20)
plt.figure(figsize=(10,20))
plt.show
```

### 긍정적 리뷰와 부정적 리뷰 파이 차트

```python
labels = {1:'긍정적 리뷰', 0:'부정적 리뷰'}
colors = ['#0984e3', '#ff7675']

sentiment = reviews.sentiment.map(labels).value_counts()
plt.figure(figsize=(10,10))
plt.pie(sentiment, labels=sentiment.index, autopct='%1.1f%%', colors=colors, textprops={'fontsize': 15})
plt.title('긍정적 리뷰와 부정적 리뷰의 비율',fontsize=20)
plt.savefig('./img/sentiment_ratio.png')
# plt.show()
```

### 전체 리뷰 각 영역별 평점 비교 

```python
durability_mean = reviews.durability_star.mean()
price_mean = reviews.price_star.mean() 
design_mean = reviews.design_star.mean() 
delivery_mean = reviews.delivery_star.mean() 

print('내구성 평균 평점:', durability_mean)
print('가격 평균 평점:', price_mean)
print('디자인 평균 평점:', design_mean)
print('배송 평균 평점:', delivery_mean)
```

```python
x = ['내구성', '가격', '디자인', '배송']
y = [durability_mean, price_mean, design_mean, delivery_mean]

tick_size = 13 
axis_label_size = 15 
 
fig = plt.figure(figsize=(10,10)) 
ax = fig.add_subplot() 
colors = '#0984e3'

xtick_label_position = list(range(len(x)))

plt.yticks(fontsize=tick_size)
plt.xticks(xtick_label_position, x, fontsize=tick_size)

bars = plt.bar(xtick_label_position, y, color=colors, 
               edgecolor=colors, alpha=0.7, linewidth=3)

for index, bar in enumerate(bars):
    ax.text(bar.get_x()+bar.get_width()*(1/2),bar.get_height()+0.05, \
            round(y[index], 2),ha='center',fontsize=13)
plt.title('전체 리뷰 하위 영역 평점',fontsize=20)
plt.xlabel('평점 영역', fontsize=axis_label_size)
plt.ylabel('평균 평점', fontsize=axis_label_size)
plt.savefig('./img/total_detail_star.png')
# plt.show()
```
