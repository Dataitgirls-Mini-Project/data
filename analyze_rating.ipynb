```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
reviews = pd.read_csv('./data/reviews_preprocessed.csv')

print(reviews.shape)
```

```python
path = '/Library/Fonts/NanumBarunGothic.otf'
font_name = fm.FontProperties(fname=path, size=50).get_name()
print(font_name)
plt.rc('font', family=font_name)
```

### 별점 히스토그램 

```python
reviews.hist(column = 'total_star', grid=False, figsize=(12,8), color='#86bf91')

plt.title("")
plt.xlabel("평점", labelpad=20, weight='bold', size=20)
plt.ylabel("리뷰 수", labelpad=20, weight='bold', size=20)

plt.show
```

### 긍정적 리뷰와 부정적 리뷰 파이 차트

```python
reviews.sentiment.value_counts().plot.pie(labels=['긍정','부정'],autopct='%1.1f%%')
plt.ylabel("")
plt.show()
```

### 각 영역별 별점 비교 

```python
durability_star_mean = reviews.durability_star.mean()
price_star_mean = reviews.price_star.mean() 
design_star_mean = reviews.design_star.mean() 
delievery_star_mean = reviews.delievery_star.mean() 

print('내구성 평균 별점:', durability_star_mean)
print('가격 평균 별점:',price_star_mean)
print('디자인 평균 별점:',design_star_mean)
print('배송 평균 별점:',delievery_star_mean)
```
