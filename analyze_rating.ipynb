```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
```

```python
reviews = pd.read_csv('./data/reviews_preprocessed.csv')

print(reviews.shape)
```

```python
path = '/Library/Fonts/NanumBarunGothic.otf'
font_name = fm.FontProperties(fname=path, size=50).get_name()
plt.rc('font', family=font_name)
```

### 전체 리뷰 별점 히스토그램 

```python
reviews.hist(column = 'total_star', grid=False, figsize=(12,8), color='#86bf91')

plt.title("")
plt.xlabel("평점", labelpad=20, weight='bold', size=20)
plt.ylabel("리뷰 수", labelpad=20, weight='bold', size=20)
plt.figure(figsize=(10,20))
plt.show
```

### 긍정적 리뷰와 부정적 리뷰 파이 차트

```python
colors1 = ['#3D7EA6','#FFA5A5']
colors2 = ['#A7BBC7','#DA7F8F']
colors3 = ['#0984e3', '#ff7675']
```

```python
reviews.sentiment.value_counts().plot.pie(labels=['3점 초과 5점 이하','3점 이하'], autopct='%1.1f%%', subplots=True, colors=colors1)
plt.ylabel("")
plt.figure(figsize=(10,10))
plt.show()
```

```python
reviews.sentiment.value_counts().plot.pie(labels=['3점 초과 5점 이하','3점 이하'], autopct='%1.1f%%', subplots=True, colors=colors2)
plt.ylabel("")
plt.figure(figsize=(10,10))
plt.show()
```

```python
reviews.sentiment.value_counts().plot.pie(labels=['3점 초과 5점 이하','3점 이하'], autopct='%1.1f%%', subplots=True, colors=colors3)
plt.ylabel("")
plt.figure(figsize=(10,10))
plt.show()
```

### 전체 리뷰 각 영역별 별점 비교 

```python
durability_mean = reviews.durability_star.mean()
price_mean = reviews.price_star.mean() 
design_mean = reviews.design_star.mean() 
delivery_mean = reviews.delivery_star.mean() 

print('내구성 평균 별점:', durability_mean)
print('가격 평균 별점:', price_mean)
print('디자인 평균 별점:', design_mean)
print('배송 평균 별점:', delivery_mean)
```

```python
plt.bar(['내구성', '가격', '디자인', '배송'],[durability_mean, price_mean, design_mean, delivery_mean])
plt.xlabel('평점 영역')
plt.ylabel('평균 평점')
plt.figure(figsize=(10,10))
plt.show()
```

### 부정적 리뷰 각 영역별 평점

```python
negative_reviews = reviews[reviews['sentiment'] == 0] 
print(negative_reviews.shape)
```

```python
plt.bar(['내구성', '가격', '디자인', '배송'],[negative_durability_mean, negative_price_mean, negative_design_mean, negative_delivery_mean])
plt.xlabel('평점 영역')
plt.ylabel('평균 평점')
plt.figure(figsize=(10,10))
plt.show()
```
